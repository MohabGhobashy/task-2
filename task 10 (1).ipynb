```python
#task 3

import requests
from bs4 import BeautifulSoup
import csv
#task 3
csvfile=open('csvfile.csv','w',encoding='utf-8',)
writer=csv.writer(csvfile)
writer.writerow(['title','date','author','description' ,'yt link'])
url='https://coreyms.com/'
website=requests.get(url).text
soup=BeautifulSoup(website,'lxml')
articles_title=soup.find_all("article")
for i in articles_title:
    writer.writerow([i.h2.text,'','', '',''])
    
article_date=soup.find_all('time',class_="entry-time")

for i in article_date:
    writer.writerow(['',i.text,'','',''])
article_author=soup.find_all('span',class_="entry-author")

for i in article_author:
    writer.writerow(['','',i.text,'',''])
article_description=soup.find_all('div',class_="entry-content")
for i in article_description:
    writer.writerow(['','','',i.p.text,''])
articles = soup.find_all('article')

for i in articles :
    headline = i.h2.a.text
    summary = i.find('div', class_ ='entry-content').p.text
    link = "None"
    
    try:
        vid = i.find('iframe', class_ = "youtube-player")['src']
        vid_id = vid.split('/')
        video10 = vid_id[4].split('?')
        c = video10[0]
        link = f'https://youtube.com/watch?v={c}'
    except Exception as e:
        link = "No Link"
        
    writer.writerow(['', '' ,'','', link ])


import pandas as pd
import numpy as np

a = pd.read_csv('csvfile.csv')

print(a)

```
